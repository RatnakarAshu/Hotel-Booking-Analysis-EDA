### Variables Description

Each column informations are given below:

1. **hotel** : Name of hotel ( City Hotel or Resort Hotel)

2. **is_cancelled** : Whether the booking is canceled or not ('1' for "cancelled" and '0' for "not cancelled")

3. **lead time** : Number of days that elapsed between the entering of date of the booking into the PMS and the arrival date.

4. **arrival_date_year** : Year of arrival date

5. **arrival_date_month** : Month of arrival date

6. **arrival_date_week_number** : Week number for arrival date

7. **arrival_date_day_of_month** : Day of arrival date

8. **stays_in_weekend_nights** : Number of weekend nights (Saturday or Sunday) the guest stayed or booked to stay at the hotel

9. **stays_in_week_nights** : Number of week nights (Monday to Friday) the guest stayed or booked to stay at the hotel

10. **adults** : Number of adults

11. **children** : Number of children

12. **babies** : Number of babies

13. **meal** : Kind of meal opted for

14. **country** : Country code

15. **market_segment** : Via which segment booking was made by the customer

16. **distribution_channel** : How the customer accessed the stay (corporate booking/Direct/TA.TO)

17. **is_repeated_guest** : Guest coming for first time or not (1 for 'Yes', 0 for 'No')

18. **previous_cancellations** : Count of previous booking cancelations

19. **previous_bookings** : Count of previous bookings

20. **reserved_room_type** : Type of room reserved by the customer

21. **assigned_room_type** : Type of room assigned to the customer

22. **booking_changes** : Count of changes made to booking

23. **deposit_type** : Deposit type at the time of making a booking (No deposit/ Non refund/ Refundable)

24. **agent** : Id of agent for booking

25. **company** : Id of company that made the booking

26. **days_in_waiting_list** : Number of days in waiting list

27. **customer_type** : Type of customer (Contract/ Group/ Transient/ Transient-Party)

28. **adr** : Average daily rate

29. **required_car_parking-spaces** : Number of car parking required

30. **total_of_special_requests** : Number of additional special requirements

31. **reservation_status** : Status of reservation (Whether a customer has checked out or canceled,or not showed)

32. **reservation_status_date** : Date of the reservation status

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Handling missing values:
# to check %age of null rows out of total rows across each columns so that we can easily manage it by removing/replacing by appropriate values
missing_value_df = missing_value.head(4).reset_index().rename(columns={'index':'column name', 0:'no. of null values'})
missing_value_df['percentage of null'] = round(100 * missing_value_df['no. of null values']/hotel_df.shape[0], 3)
missing_value_df
````

````python
# delet column 'company' because it contain 93.98% null values and not contribute alot in the analysis
hotel_df = hotel_df.copy()
hotel_df.drop('company', axis=1, inplace=True)
hotel_df.shape[1]
````

````python
# replace all null value in the column "agent" by "0" because it contain only 13.95% null values so it may have good contribution in the analysis

hotel_df['agent'].fillna(0, inplace=True)
````

````python
# replace all null value in the column "country" by "others" because it contain only 0.51% null values so it has good contribution in the analysis

hotel_df['country'].fillna('other', inplace=True)
````

````python
# replace all null value in the column "children" by mean of it because it has only 0.005% of null values so it has good contribution in the analysis

hotel_df['children'].fillna(hotel_df['children'].mean(), inplace=True)
````

````python
hotel_df.isna().sum()
````

## Data Cleaning

````python
# Convert data types of some columns

hotel_df[['children', 'agent']] = hotel_df[['children', 'agent']].astype(int)     # Convert Column 'children', 'agent' from float to integer data type
hotel_df['reservation_status_date'] = pd.to_datetime(hotel_df['reservation_status_date'])
````

````python
hotel_df.info()
````

````python
# creating 2 new columns "total_guest" and "revenue"
hotel_df['total_guest'] = hotel_df['adults'] + hotel_df['children'] + hotel_df['babies']    # This column will help us to know the total people stay and revenue of each hotel
hotel_df['revenue'] = hotel_df['adr'] * hotel_df['total_guest']    # This column will help us to analyse the profit/growth of each hotel

````

#### Chart - 5    What is the trend of booking with in a year?

````python
# extract information of total booking of hotels in each month and rename the column "hotel" to 'total_booking' and save it in new data frame as monthly_booking
monthly_booking = booking_not_canceled.groupby('arrival_date_month')['hotel'].count().reset_index().rename(columns={'hotel':'total_booking'})  # for those the booking not canceled by the guest.

# create list of month in order
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
monthly_booking['arrival_date_month'] = pd.Categorical(monthly_booking['arrival_date_month'], categories=month_order, ordered=True)
monthly_booking = monthly_booking.sort_values('arrival_date_month')

# set the plot size
plt.figure(figsize=(15,5))
# create line plot
plt.plot(monthly_booking['arrival_date_month'], monthly_booking['total_booking'], linestyle='dashdot', marker='o', color='#793FDF')
#set labels and title for plot
plt.xlabel('Months', size=12)
plt.ylabel('Total Booking', size=12)
plt.title('Total booking in each month', size=15, color='#793FDF')
plt.show()
````
